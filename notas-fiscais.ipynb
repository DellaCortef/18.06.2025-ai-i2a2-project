```python
# importing libs
import os
import zipfile
import pandas as pd
```

```python
# Caminho atual onde o Jupyter está rodando
current_dir = os.getcwd()
print(f"Caminho atual: {current_dir}")
```

```python
# Define os caminhos com base no diretório atual
zip_path = os.path.join(current_dir, "Archive.zip")
extract_dir = os.path.join(current_dir, "extracted")

# Cria a pasta de extração se não existir
os.makedirs(extract_dir, exist_ok=True)

# Descompacta o arquivo
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Arquivos extraídos:")
print(os.listdir(extract_dir))
```

```python
# Caminhos completos para os arquivos CSV
cabecalho_path = os.path.join(extract_dir, "202401_NFs_Cabecalho.csv")
itens_path = os.path.join(extract_dir, "202401_NFs_Itens.csv")

# Leitura dos arquivos
df_cabecalho = pd.read_csv(cabecalho_path)
df_itens = pd.read_csv(itens_path)

# Verifica rapidamente se foi carregado
print("Cabecalho:")
print(df_cabecalho.shape)
print("Itens:")
print(df_itens.shape)
```

```python
# Conversão no df_cabecalho
df_cabecalho["DATA EMISSÃO"] = pd.to_datetime(df_cabecalho["DATA EMISSÃO"])
df_cabecalho["DATA/HORA EVENTO MAIS RECENTE"] = pd.to_datetime(df_cabecalho["DATA/HORA EVENTO MAIS RECENTE"])

# Se df_itens também tiver datas, por exemplo:
if "DATA EMISSÃO" in df_itens.columns:
    df_itens["DATA EMISSÃO"] = pd.to_datetime(df_itens["DATA EMISSÃO"])
```
